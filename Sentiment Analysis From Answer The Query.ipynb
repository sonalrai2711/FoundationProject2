# Sentiment Analysis - Energy Consumption

```python
#Data processing packages
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', 200)

#Visualization packages
import matplotlib.pyplot as plt
import seaborn as sns

#NLP packages
from textblob import TextBlob

import warnings
warnings.filterwarnings("ignore")
```

```python
#Testing NLP - Sentiment Analysis using TextBlob
TextBlob("Energy is bad in San Diego").sentiment
```

```python
questions = pd.read_csv('C:/Users/chitr/OneDrive/Desktop/ISB/Foundation Project 2/Energy Project/Sentiment Analysis/San_Diego_Energy_Consumption.csv',encoding='utf8',error_bad_lines=False);#opening the file UScomments
questions.head()
```

```python
#Calculating the Sentiment Polarity
pol=[] # list which will contain the polarity of the comments
for i in questions.Suggestion.values:
    try:
        analysis =TextBlob(i)
        pol.append(analysis.sentiment.polarity)
        
    except:
        pol.append(0)
```

```python
questions['polarity']=pol
```

```python
questions['polarity'][questions.polarity==0]= 0
questions['polarity'][questions.polarity > 0]= 1
questions['polarity'][questions.polarity < 0]= -1 
```

```python
df_positive = questions[questions.polarity==1]
df_positive.head(10)
```

```python
df_negative = questions[questions.polarity==-1]
df_negative.head()
```

```python
df_neutral = questions[questions.polarity==0]
df_neutral.head(10)
```

```python
questions.polarity.value_counts().plot.bar()
questions.polarity.value_counts()
```

```python
all_words = ' '.join([text for text in questions['Suggestion']])
from wordcloud import WordCloud
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
